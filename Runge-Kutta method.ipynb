```python
#Рунге-Кутта 4
import math
import numpy as np
import matplotlib.pyplot as plt
import sys

def runge_kutta4(y, t0, h, f):
    
    k1 = h*f(y, t0)
    k2 = h*f(y+k1/2, t0+h/2)
    k3 = h*f(y+k2/2, t0+h/2)
    k4 = h*f(y+k3, t0+h)
    return y+k1/6+k2/3+k3/3+k4/6.

def runge_kutta2(y, t0, h, f):
    k1 = h*f(y, t0)
    k2 = h*f(y+k1/2, t0+h/2)
    return y+k1/6+k2

t = 0.
x = 0.
dt = 0.01
t1 = 0.
x1= 0.
dt1 = 0.01
xs, ts = [], []
xss, tss = [], []
def func(x, t):
    return t+2*x*t

while t <= 2:
    x = runge_kutta4(x, t, dt, func)
    t += dt
    xs.append(x)
    ts.append(t)
while t1 <= 2:
    x1 = runge_kutta2(x1, t1, dt1, func)
    t1 += dt1
    xss.append(x1)
    tss.append(t1)
exact = [1./2.*((math.e**(t)**2)-1.) for t in ts]
plt.plot(ts,xs,  label='Рунге-Кутта 4')
plt.legend()
plt.show()
```

```python
#Рунге-Кутта 4
import math
import numpy as np
import matplotlib.pyplot as plt
import sys

def runge_kutta4(y, t0, h, f):
    
    k1 = h*f(y, t0)
    k2 = h*f(y+k1/2, t0+h/2)
    k3 = h*f(y+k2/2, t0+h/2)
    k4 = h*f(y+k3, t0+h)
    return y+k1/6+k2/3+k3/3+k4/6.
t = 0.
x = 0.
dt = 0.1
N0=10**12
N=np.linspace(0,100,100) 
u=18 #km/ch
Ns=[]
for i in range(0,100,5):
    N[i]= N0*math.e**(-((i-50)**2)/8)
    
    Ns.append(N[i])
xs, ts = [], []
xss, tss = [], []
def func(x, t):
    return (-u*Ns)

while t <= 2:
    x = runge_kutta4(x, t, dt, func)
    t += dt
    xs.append(x)
    ts.append(t)
    
plt.plot(ts,xs,  label='Рунге-Кутта 4')
plt.legend()
plt.show()
```
